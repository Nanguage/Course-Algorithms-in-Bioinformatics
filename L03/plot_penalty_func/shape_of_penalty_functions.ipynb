```python
import math
```

```python
def logistic(x, x0=0, a=1, b=math.e, c=1, L=1):
    return L / (a + b**(-c*(x-x0)))
```

```python
def plot_logistic_func(X, a, b, c, N=100):
    X = np.linspace(0, N)
    func = lambda x: logistic(x, 0, a, b, c, 1)
    Y = np.apply_along_axis(func, 0, X)
    plt.plot(X, Y)
    plt.xlim(0)
    plt.ylim(0, 1.1)
```

```python
plot_logistic_func(X, 1, math.e, 1, N=10)
```
